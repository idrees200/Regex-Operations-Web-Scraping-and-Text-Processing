```python
import nltk
import re
def test_regex(regexp, text):
    print(f"Regular Expression: {regexp}")
    print(f"Text: {text}")
    nltk.re_show(regexp, text)
    print("\n" + "-"*50 + "\n")
```

```python
test_regex(r'[a-zA-Z]+', "Hello World 123")
test_regex(r'[A-Z][a-z]*', "Variable Name")
test_regex(r'p[aeiou]{,2}t', "pat pet pit pot put")
test_regex(r'\d+(\.\d+)?', "123 3.14 0.5")
test_regex(r'([^aeiou][aeiou][^aeiou])*', "cat dog owl")
test_regex(r'\w+|[^\w\s]+', "word123 !@#$")
```

```python
from urllib import request
from bs4 import BeautifulSoup
import re
```

```python
def fetch_url_contents(url):
    response = request.urlopen(url)
    html = response.read().decode('utf8')
    soup = BeautifulSoup(html, 'html.parser')
    text = soup.get_text()
    return text
```

```python
def extract_phone_numbers_and_emails(text):
    phone_numbers = re.findall(r'(\b(?:\+\d{1,2}\s?)?(\d{3}[-.\s]?){2}\d{4}\b)', text)
    email_addresses = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b', text)
    return phone_numbers, email_addresses
```

```python
url = 'https://www.nu.edu.pk/University/PhDFaculty'
text = fetch_url_contents(url)
phone_numbers, email_addresses = extract_phone_numbers_and_emails(text)
```

```python
print("Phone numbers:")
for phone_number in phone_numbers:
    print(phone_number)
print("\nEmail addresses:")
for email_address in email_addresses:
    print(email_address)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
```

```python
def normalize_text(text, stemmer):
    tokens = nltk.word_tokenize(text)
    normalized_tokens = [stemmer.stem(token) for token in tokens]
    return normalized_tokens
```

```python
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
```

```python
normalized_tokens_porter = normalize_text(text, porter_stemmer)
normalized_tokens_lancaster = normalize_text(text, lancaster_stemmer)
```

```python
print("Porter Stemmer:")
for token in normalized_tokens_porter:
    print(token)
```

```python
print("\nLancaster Stemmer:")
for token in normalized_tokens_lancaster:
    print(token)
```

```python
first_word_pattern = re.compile(r'[Zz][a-z]*a')
first_word_matches = first_word_pattern.findall(text)
first_word_count = len(first_word_matches)
first_word = first_word_matches if first_word_count > 0 else None
```

```python
print("a) Regular Expression:", r'[Zz][a-z]*a')
print("b) Frequency (count):", first_word_count)
print("c) First Word:", first_word)
```

```python
regex = r'\b\d[kK][a-zA-Z]*\d\b'
ex = re.search(regex,text).group(0)
second_word = ex[3:-3]
print("Regular Expression:",ex)
print("second word:", second_word)
```

```python
third_word = '[c][a-z][*][a-z][a]'
tw = re.findall(third_word, text)
third = tw[0].replace('*', '').replace('[', '').replace(']', '')
print("Regular expression:",tw)
print("Third word:",third)
```

```python
fourth = r'[Pp][a-z]{2}[a]'
frequency = len(re.findall(fourth, text))
fourth_word = re.search(fourth, text).group(0)
print("Fourth word:", fourth_word)
print("Frequency of fourth word:",frequency)
```
